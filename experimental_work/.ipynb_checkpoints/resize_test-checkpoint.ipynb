```python
from PIL import Image
from torchvision import transforms 
from torchvision.transforms import Compose, Resize, CenterCrop, ToTensor, Normalize
import os
import tqdm

def main(path, save_path):
    preprocess = transforms.Compose([
            Resize(256, interpolation=Image.BICUBIC),
    ])
    openimages_imgs = os.listdir(path)
    for index, img in enumerate(tqdm.tqdm(openimages_imgs)):
        im = Image.open(path+img)
        img_transformed = preprocess(im)
        img_transformed.save(save_path+img)    
```

```python
openimages_imgs = os.listdir("oi_test/")
preprocess = transforms.Compose([
        Resize((224, 224), interpolation=Image.BICUBIC),
        CenterCrop(224),
        ToTensor()
])
for i in openimages_imgs:
    im = Image.open("oi_test/"+i).
    print(im.size)
    img = preprocess(im)
    
```

```python
main("../VisionResearch/finetuneClip/openimages_dataset/val/", "oi_test/")
```
